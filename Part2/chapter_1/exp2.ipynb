```python
import cv2
import matplotlib.pyplot as plt
```

multiplt image visualize 그리고, figure 와 axs 사용.

```python
im_name = '../chapter_1/Chapter_1_source_code/Chapter_1/data/image.jpeg'
image = cv2.imread(im_name)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(image)
plt.show()
plt.imshow(image_rgb)
plt.show()
plt.imshow(image_gray, cmap='gray')
plt.show()
```

```python
plt.figure(figsize=(15,8))

plt.subplot(1,3,1)
plt.imshow(image)
plt.axis('off')
plt.title('image1')

plt.subplot(1,3,2)
plt.imshow(image_rgb)
plt.axis('off')
plt.title('image_rgb')

plt.subplot(1,3,3)
plt.imshow(image_gray, cmap='gray')
plt.axis('off')

plt.show()
```
