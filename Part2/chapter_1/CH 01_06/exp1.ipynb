# Loading data

* Methods you should define when inherit from torch.utils.Dataset
1. __init__(self,[]) : 데이터셋 루트 디렉토리 / transform / 이미지 path ...
2. __len__(self) : 데이터셋의 length(크기) 반환
3. __getitem__(self, idx) : (idx)th data를 return. *Dataloader call this method
