```python
import torch, torchvision
import torchvision.models as models

import matplotlib.pyplot as plt
from PIL import Image
```

```python
models.vgg16()
```

```python
cifar10 = torchvision.datasets.CIFAR10(root='./', download=True)

to_tensor = torchvision.transforms.Compose([
    torchvision.transforms.ToTensor(),
    torchvision.transforms.Normalize(mean=[0.485, 0.456, 0.406],
                                     std=[0.229, 0.224, 0.225])
])

vgg16 = models.vgg16(pretrained=True)
```

```python
for img in cifar10:
    img = to_tensor(img[0])
    print(img.shape)

    logit = vgg16(img.unsqueeze(0))

    print(logit.shape)
    break
```

# Make dataloader
see the document!  
https://pytorch.org/docs/stavle/data.html

```python
cifar10 = torchvision.datasets.CIFAR10(root='./', download=True, transform=to_tensor)

dataloader = torch.utils.data.DataLoader(cifar10, batch_size=8, shuffle=True, num_workers=2)
```

```python
dataloader
```

```python
for idx, data in enumerate(dataloader):
    
    img, gt = data

    print(img.shape, gt.shape)

    # for im in img:
    #     plt.imshow(torchvision.transforms.ToPILImage()(im))
    #     plt.show()

    # TODO: run the vgg16 network
    logits = vgg16(img)

    print(logits.shape)

    break
```
